loc와 iloc는 약간 다르다.

loc -> index name기반으로 추출하고 pandas에 가깝

iloc -> index 기반으로 한다. numpy에 가깝

```python
tips.loc[3,'day']
```

```python
tips.loc[3:6,'day']
```

```python
tips.loc[3:6,['day','smoker']]
```

```python
tips.loc[3:6,['tip','smoker']]
```

```python
tips.loc[3:6,['smoker','tip']]
```

```python
tips.iloc[3:6,:]
```

```python
tips.iloc[3:6,::-1]
```

```python
tips.iloc[3:6,...]
```
