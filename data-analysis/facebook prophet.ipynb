```python
import pandas as pd
```

```python
# https://raw.githubusercontent.com/facebook/prophet/master/examples/example_wp_log_peyton_manning.csv
from fbprophet import Prophet
# Prophet 클래스 -> 인스턴스해서 사용한다.
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/facebook/prophet/master/examples/example_wp_log_peyton_manning.csv')
```

```python
data
```

```python
m = Prophet
# 인스턴스화
```

```python
# 비교를 해보자
vars(m)
```

```python
m.fit(data)
# 학습시키기
```

```python
vars(m)
```

```python
# datatime timedelta period
f = m.make_future_dataframe(365)
```

```python
f
```

```python
vars(m)
```

```python
# 예측하기
forecast = m.predict(f)
```

```python
# 팬시 우우우~
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
# Rodeo spider에러많은걸 대응하기위한 분석 전용 ide rodeo
```

```python
fig1 = m.plot(forecast)
```

```python
fit2 = m.plot_components(forecast)
```
