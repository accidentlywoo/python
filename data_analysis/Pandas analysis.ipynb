pandas에서는 어그리게이션 분석을 지원한다. 

mean, sum, max ...

```python
tips.groupby('sex').mean()
```

```python
tips.groupby('sex').sum()
```

```python
tips.groupby('day').max()
```

```python
tips.groupby(['day','sex']).max()
```

```python
tips.groupby(['day','sex'])['total_bill','tip'].max()
```

```python
tips.groupby(['sex','day','time'])['total_bill','tip'].max()
```
